```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv =  path + '/zomato.csv'
```

```python
df.drop(['phone','dish_liked','url' , 'menu_item'],axis=1,inplace=True)
```

```python
df.rename(columns={'approx_cost(for two people)': 'cost', 'listed_in(type)': 'service_type'}, inplace=True)
```

```python
df = df.dropna()
```

```python
df['cost'] = df['cost'].str.replace(',', '').astype(float)
```

```python
plt.figure(figsize=(10,6))
df['location'].value_counts().head(10).plot(kind='bar', color='tomato')
plt.title("Top 10 Restaurant Locations")
plt.xlabel("Location")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['cuisines'].value_counts().head(10).plot(kind='barh', color='orchid')
plt.title("Top 10 Cuisines")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisines")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(78,35))
sns.histplot(df['rate'], bins=20, kde=True, color='red')
plt.title("Distribution of Restaurant Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(15,60))
sns.scatterplot(x='cost', y='rate', data=df, hue='online_order', palette='viridis')
plt.title("Price vs Rating")
plt.xlabel("Cost for Two")
plt.ylabel("Rating")
plt.show()
```

```python
plt.figure(figsize=(6,6))
df['online_order'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen'])
plt.title("Online Order Availability")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(107,9))
df['service_type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['gold', 'lightskyblue'])
plt.title("Dine-in vs Delivery Types")
plt.ylabel("")
plt.show()
```

```python
def rating_category(r):
    if pd.isna(r):
        return 'Unknown' # Handle potential NaN values in 'rate'
    # Convert rate to a numerical value before comparison
    try:
        rate_value = float(r.split('/')[0])
    except (ValueError, AttributeError):
        return 'Unknown' # Handle non-numeric or unexpected formats
    if rate_value < 3:
        return 'Low'
    elif rate_value < 4:
        return 'Medium'
    else:
        return 'High'

df['rating_category'] = df['rate'].apply(rating_category)

# Display the counts of each rating category
display(df['rating_category'].value_counts())
```
